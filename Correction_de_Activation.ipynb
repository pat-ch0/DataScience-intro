```python
import math
w1 = .4
w2 = .6
w3 = .1
i1 = .1
i2 = .5
out = .2
eta = .5

def sigmoid(x):
  s = 1.0 / (1.0 + math.exp(-x))
  return s
```

*  `out` : Cette variable représente la sortie souhaitée ou la valeur cible que le réseau neuronal est entraîné à prédire.
`eta` : Cette variable représente le taux d'apprentissage. Le taux d'apprentissage contrôle la façon dont les poids sont ajustés au cours de chaque itération d'entraînement.

*  La fonction `sigmoid` prend une seule entrée (`x`) et calcule une valeur en utilisant la formule sigmoïde : `1 / (1 + exp(-x))`.

Enfin, la valeur calculée (`s`) est renvoyée par la fonction.
**En substance, cette fonction `sigmoid` est une fonction mathématique couramment utilisée dans les réseaux neuronaux comme fonction d'activation.** Les fonctions d'activation introduisent une non-linéarité dans le réseau, lui permettant d'apprendre des schémas complexes. La fonction sigmoïde mappe spécifiquement toute valeur d'entrée à une valeur comprise entre 0 et 1, souvent interprétée comme une probabilité.

## Passage en avant (Forward Pass)

Cette partie du code simule la manière dont le réseau neuronal traite les données d'entrée pour produire une sortie.

```python
# forward
net = w1 * i1 * i1 + w2 * i2
outh = sigmoid(net)
neto = w3 * outh
o = sigmoid(net)
```

1. **Calculer la somme pondérée des entrées pour la couche cachée :**

  *  `net = w1 * i1 * i1 + w2 * i2`
    *  Cette ligne calcule la somme pondérée des entrées (`i1`, `i2`) multipliées par leurs poids correspondants (`w1`, `w2`).
    *  Cette somme pondérée (`net`) représente l'entrée totale du neurone de la couche cachée.

2.  **Appliquer la fonction d'activation à la couche cachée :**

  *  `outh = sigmoid(neth)`
     *  La fonction sigmoid est appliquée à la somme pondérée des entrées (en supposant que 'neth' est 'net') calculée à l'étape précédente.
     *  Cela introduit une non-linéarité dans le réseau. La sortie de la fonction sigmoid (outh) représente l'activation du neurone de la couche cachée.

3.  **Calculer la somme pondérée des entrées pour la couche de sortie** :

  *  `neto = w3 * outh`
     *  Cette ligne calcule la somme pondérée de la sortie de la couche cachée (`outh`) multipliée par le poids (`w3`).
     *  Cette somme pondérée (`neto`) représente l'entrée totale du neurone de la couche de sortie.

4.  **Appliquer la fonction d'activation à la couche de sortie** :

  *  `o = sigmoid(net)`
     *  La fonction `sigmoid` est appliquée à l'entrée totale de la couche de sortie (`neto`).
     *  La sortie de la fonction `sigmoid (o)` représente la sortie finale du réseau neuronal.


En termes plus simples, ce code prend deux entrées (`i1`, `i2`), les traite via un neurone de couche cachée en utilisant des poids et la fonction d'activation sigmoïde, puis génère une sortie finale (`o`) via un neurone de couche de sortie. Ce processus de données circulant de l'entrée à la sortie est appelé le passage en avant (forward pass).

## Backward Pass

Le code suivant effectue les étapes suivantes pour chaque itération :
Backpropagation-passes-architecture.webp
1.  Calcule la sortie du réseau pour les entrées données (**Forward Pass**).
2.  Calcule l'erreur entre la sortie du réseau et la sortie souhaitée.
3.  Commence à calculer les ajustements à apporter aux poids pour réduire l'erreur (**début du Backward Pass**).

L'objectif final de la rétropropagation, qui n'est que partiellement implémentée ici, est de propager l'erreur en arrière à travers le réseau et d'ajuster les poids de manière à minimiser cette erreur au fil des itérations. Ceci est fait en utilisant les dérivées partielles de l'erreur par rapport à chaque poids, comme illustré par le calcul de `dw3` et le début du calcul de `dw2`.

```python
for i in range(1000):
  neth = w1 * i1 + w2 * i2
  outh = sigmoid(neth)
  neto = w3 * outh
  o = sigmoid(neto)
  err = 0.5 * (out - o) ** 2
  print(err, o)
  dw3 = (o-out) * o * (1.0 - o) * outh
  dw2 = (o-out)
```

Ce code représente une partie du processus d'apprentissage d'un réseau neuronal simple, plus précisément la **phase de rétropropagation (backward pass)** et le calcul de l'erreur.

1.  **Boucle d'apprentissage** `for i in range(1000):` initie une boucle qui va s'exécuter 1000 fois. Chaque itération de cette boucle représente une étape d'apprentissage du réseau.

2. **Passage en avant (Forward Pass)**:

  *  `neth = w1 * i1 + w2 * i2` : Calcul de l'entrée `neth` du neurone caché en utilisant les poids `w1`, `w2` et les entrées `i1`, `i2`.
  *  `outh = sigmoid(neth)` : Application de la fonction d'activation `sigmoid` à `neth` pour obtenir la sortie `outh` du neurone caché.
  *  `neto = w3 * outh` : Calcul de l'entrée `neto` du neurone de sortie en utilisant le poids `w3` et la sortie du neurone caché `outh`.
  *  `o = sigmoid(neto)` : Application de la fonction d'activation `sigmoid` à `neto` pour obtenir la sortie finale `o` du réseau.

3. **Calcul de l'erreur** :

  *  `err = 0.5 * (out - o) ** 2` calcule l'erreur du réseau.
    *  `out` représente la sortie souhaitée (la valeur cible).
    *  `o` représente la sortie réelle du réseau.
    *  La formule utilisée est celle de l'**erreur quadratique moyenne**, divisée par 2 pour des raisons de simplification mathématique lors de la rétropropagation.

4.  **Affichage de l'erreur et de la sortie** :

  *  `print(err, o)`: Affiche l'erreur `err` et la sortie du réseau `o` à chaque itération. Cela permet de suivre l'évolution de l'apprentissage.
5.  **Début du calcul des gradients (Rétropropagation - Backward Pass)** :

  *  `dw3 = (o-out) * o * (1.0 - o) * outh` calcule la dérivée partielle de l'erreur par rapport au poids `w3`. Ce calcul est basé sur la règle de la chaîne et utilise la dérivée de la fonction `sigmoid`. C'est une étape cruciale de la rétropropagation, car elle permet de déterminer comment ajuster `w3` pour réduire l'erreur.
  *  `dw2 = (o-out)` est un début de calcul de la dérivée partielle de l'erreur par rapport à `w2`, mais elle est incomplète dans le code fourni. Elle devrait également inclure des termes liés à la dérivée de la fonction `sigmoid` et aux autres poids/activations du réseau.

  Backpropagation-prediction.webp

```python
for i in range (1000):
    neth = w1 * i1 + w2 * i2
    outh = sigmoid (neth)
    neto = w3 * outh
    o = sigmoid (neto)
    err = 0.5 * (out - o) ** 2
    print (err, o)
    dw3 = (o - out) * o * (1.0 - o) * outh
    dw2 = (o - out) * o * (1.0 - o) * w3 * outh * (1.0 - outh) * i2
    dw1 = (o - out) * o * (1.0 - o) * w3 * outh * (1.0 - outh) * i1
    w3 = w3 - eta * dw3
    w2 = w2 - eta * dw2
    w1 = w1 - eta * dw1
```

# Cross entropy loss

```python
# binary cross entropy
out = 1.0
for i in range(1000):
  neth = w1 * i1 + w2 * i2
  outh = sigmoid(neth)
```

**Ce code initialise un processus d'apprentissage pour un réseau neuronal simple utilisant l'entropie croisée binaire comme fonction de coût.**

*  **Fonction de coût** : L'entropie croisée binaire est utilisée pour mesurer l'erreur entre la prédiction du réseau et la valeur cible. Elle est particulièrement adaptée aux problèmes de classification binaire.
*  **Boucle d'apprentissage**: La boucle `for` permet au réseau d'apprendre en ajustant ses poids à chaque itération pour minimiser l'erreur.
*  **Passage en avant (Forward Pass)**: Les lignes `neth = ...` et `outh = ...` représentent le passage en avant, où le réseau calcule sa sortie en fonction des entrées et des poids actuels.
*  Le code ne montre que le passage en avant et le calcul de la sortie du neurone caché. Il manque la partie **rétropropagation (backward pass)** où les poids du réseau seraient ajustés en fonction de l'erreur calculée. La rétropropagation est essentielle pour l'apprentissage du réseau.

```python
# binary cross entropy
out = 1.0
for i in range (1000):
    neth = w1 * i1 + w2 * i2
    outh = sigmoid (neth)
    neto = w3 * outh
    o = sigmoid (neto)
    err = -out * math.log (o) - (1.0 - out) * math.log (1.0 - o)
    print (err, o)
    dw3 = (-out / o) * o * (1.0 - o) * outh
    dw2 = (-out / o) * o * (1.0 - o) * w3 * outh * (1.0 - outh) * i2
    dw1 = (-out / o) * o * (1.0 - o) * w3 * outh * (1.0 - outh) * i1
    w3 = w3 - eta * dw3
    w2 = w2 - eta * dw2
    w1 = w1 - eta * dw1
```

Ce code implémente un réseau neuronal simple avec une couche cachée et une couche de sortie, et il entraîne le réseau en utilisant l'entropie croisée binaire comme fonction de coût. Cela signifie que le réseau est conçu pour des tâches de classification binaire (prédire l'un des deux résultats possibles).
*  Initialisation : `out = 1.0` définit la sortie cible ou la prédiction souhaitée du réseau à `1.0`. Rappelez-vous qu'il s'agit d'un problème de classification binaire, donc la sortie cible est soit 0, soit 1.
*  L'entropie croisée binaire est une fonction de coût bien adaptée à la classification binaire car elle pénalise fortement les prédictions qui sont confiantes mais erronées. Elle encourage le réseau à produire des sorties proches de 0 ou 1, s'alignant sur la nature de la classification binaire.
*  Les gradients sont utilisés pour mettre à jour les poids (`w1`, `w2`, `w3`) dans la direction qui réduit l'erreur.
*  `eta` est le taux d'apprentissage, contrôlant la taille des ajustements de poids.

```python
# binary cross entropy
out1 = 1.0
out2 = 0.0
w4 = 0.5
for i in range (100000):
    neth = w1 * i1 + w2 * i2
    outh = sigmoid (neth)
    neto1 = w3 * outh
    o1 = sigmoid (neto1)
    neto2 = w4 * outh
    o2 = sigmoid (neto2)
    err = -out1 * math.log (o1) - (1.0 - out2) * math.log (1.0 - o2)
    print (err, o1, o2)
    do2 = (1 - out2) / (1.0 - o2) * o2 * (1.0 - o2)
    do1 = (-out1 / o1) * o1 * (1.0 - o1)
    dw4 = do2 * outh
    dw3 = do1 * outh
    dw2 = (do1 * w3 + do2 * w4) * outh * (1.0 - outh) * i2
    dw1 = (do1 * w3 + do2 * w4) * outh * (1.0 - outh) * i1
    w4 = w4 - eta * dw4
    w3 = w3 - eta * dw3
    w2 = w2 - eta * dw2
    w1 = w1 - eta * dw1
```

```python
# softmax + cross entropy
out1 = 1.0
out2 = 0.0
w4 = 0.5
for i in range (100000):
    neth = w1 * i1 + w2 * i2
    outh = sigmoid (neth)
    neto1 = w3 * outh
    neto2 = w4 * outh
    # normalization factor for the softmax function.
    sigma = math.exp (neto1) + math.exp (neto2)
    o1 = math.exp (neto1) / sigma
    o2 = math.exp (neto2) / sigma
    err = -out1 * math.log (o1) - out2 * math.log (o2)
    print (err, o1, o2)
    do2 = (o2 - out2) * o2 * (1.0 - o2)
    do1 = (o1 - out1) * o1 * (1.0 - o1)
    dw4 = do2 * outh
    dw3 = do1 * outh
    dw2 = (do1 * w3 + do2 * w4) * outh * (1.0 - outh) * i2
    dw1 = (do1 * w3 + do2 * w4) * outh * (1.0 - outh) * i1
    w4 = w4 - eta * dw4
    w3 = w3 - eta * dw3
    w2 = w2 - eta * dw2
    w1 = w1 - eta * dw1
```

La fonction softmax est une fonction d'activation utilisée dans les réseaux de neurones pour convertir les scores de sortie en probabilités. Elle prend un vecteur de valeurs (scores) en entrée et les transforme en un vecteur de probabilités de la manière suivante :

1. Calcule l'exponentielle de chaque score.
2. Calcule la somme de toutes les exponentielles.
3. Divise chaque exponentielle par cette somme.

Cela garantit que les valeurs de sortie se situent entre 0 et 1 et que leur somme est égale à 1, ce qui en fait des probabilités.
